```python
# O script que cria os arquivos .gitkeep

import os
import fnmatch

from google.colab import drive
drive.mount('/content/drive')

# CONFIGURAÇÃO: Caminho exato do seu projeto no Google Drive
PASTA_RAIZ = "/content/drive/MyDrive/UFPA/Disciplinas/2026.2/CO18008 - Arquitetura e Organização de Computadores/Artigo/aoc-empirical-benchmark"

def carregar_padroes_gitignore(diretorio_alvo):
    caminho_gitignore = os.path.join(diretorio_alvo, '.gitignore')
    padroes = ['.git'] # Ignora a pasta oculta do Git por padrão

    if os.path.exists(caminho_gitignore):
        with open(caminho_gitignore, 'r', encoding='utf-8') as f:
            for linha in f:
                linha = linha.strip()
                # Ignora linhas vazias e comentários do gitignore
                if linha and not linha.startswith('#'):
                    padroes.append(linha.rstrip('/'))
    return padroes

def esta_ignorado(caminho_relativo, padroes):
    partes = caminho_relativo.split('/')
    for parte in partes:
        for padrao in padroes:
            if fnmatch.fnmatch(parte, padrao) or fnmatch.fnmatch(f"/{parte}", padrao):
                return True
    return False

def colocar_gitkeep_com_gitignore(diretorio_alvo):
    if not os.path.exists(diretorio_alvo):
        print(f"❌ Erro: O caminho não foi encontrado. Verifique se o Google Drive está montado.")
        return

    padroes_ignorados = carregar_padroes_gitignore(diretorio_alvo)
    contador = 0

    for raiz, pastas, arquivos in os.walk(diretorio_alvo):
        caminho_relativo = os.path.relpath(raiz, diretorio_alvo)

        # Pula a pasta se ela estiver listada no .gitignore
        if caminho_relativo != '.' and esta_ignorado(caminho_relativo, padroes_ignorados):
            pastas.clear() # Evita entrar em subpastas ignoradas
            continue

        # Filtra subpastas ignoradas
        pastas[:] = [p for p in pastas if not esta_ignorado(os.path.relpath(os.path.join(raiz, p), diretorio_alvo), padroes_ignorados)]

        # Cria o .gitkeep se a pasta estiver completamente vazia
        if not pastas and not arquivos:
            caminho_gitkeep = os.path.join(raiz, '.gitkeep')
            with open(caminho_gitkeep, 'w') as f:
                pass
            print(f"Criado: {caminho_relativo}/.gitkeep")
            contador += 1

    print(f"\n✨ Pronto! {contador} arquivos .gitkeep criados no seu projeto de AOC.")

# Executa o script
colocar_gitkeep_com_gitignore(PASTA_RAIZ)
```
